```python
from random import random
from random import randint
import pandas as pd
```

```python
TotalTime = []
PictureClicks = []
PictureTime = []
PictureDwelling = []
SummaryClicks = []
SummaryTime = []
SummaryDwelling = []
DescriptionClicks = []
DescriptionTime = []
DescriptionDwelling = []
ReviewClicks = []
ReviewTime = []
ReviewDwelling = []
Return = []
```

```python
for i in range (0,5000) :
    TotalTime.append(round((random()+randint(20,79))*1000,0))
    PictureClicks.append(randint(3,5))
    PictureTime.append(round((random()+randint(5,29))*1000,0))
    PictureDwelling.append(min(round((random()+randint(5,29))*1000,0),PictureTime[i]))
    SummaryClicks.append(randint(2,3))
    SummaryTime.append(round((random()+randint(5,29))*1000,0))
    SummaryDwelling.append(min(round((random()+randint(3,9))*1000,0),SummaryTime[i]))
    DescriptionClicks.append(randint(1,3))
    DescriptionTime.append(round((random()+randint(4,19))*1000,0))
    DescriptionDwelling.append(min(round((random()+randint(4,19))*1000,0),DescriptionTime[i]))
    ReviewClicks.append(randint(2,15))
    ReviewTime.append(round((random()+randint(4,29))*1000,0))
    ReviewDwelling.append(min(round((random()+randint(4,29))*1000,0),ReviewTime[i]))
    Return.append(0)

for j in range (5000,10000) :
    TotalTime.append(round((random()+randint(3,44))*1000,0))
    PictureClicks.append(randint(0,4))
    PictureTime.append(round((random()+randint(0,19))*1000,0))
    PictureDwelling.append(min(round((random()+randint(0,19))*1000,0),PictureTime[j]))
    SummaryClicks.append(randint(3,6))
    SummaryDwelling.append(round((random()+randint(0,29))*1000,0))
    SummaryTime.append(max(round((random()+randint(0,19))*1000,0),SummaryDwelling[j]))
    DescriptionClicks.append(randint(0,2))
    DescriptionTime.append(round((random()+randint(0,14))*1000,0))
    DescriptionDwelling.append(min(round((random()+randint(0,9))*1000,0),DescriptionTime[j]))
    ReviewClicks.append(randint(0,3))
    ReviewTime.append(round((random()+randint(0,19))*1000,0))
    ReviewDwelling.append(min(round((random()+randint(0,9))*1000,0),ReviewTime[j]))
    Return.append(1)
    
for k in range (10000,12000) :
    TotalTime.append(round((random()+randint(0,79))*1000,0))
    PictureClicks.append(randint(0,5))
    PictureTime.append(round((random()+randint(0,29))*1000,0))
    PictureDwelling.append(min(round((random()+randint(0,29))*1000,0),PictureTime[k]))
    SummaryClicks.append(randint(0,6))
    SummaryTime.append(round((random()+randint(0,29))*1000,0))
    SummaryDwelling.append(min(round((random()+randint(0,9))*1000,0),SummaryTime[k]))
    DescriptionClicks.append(randint(0,5))
    DescriptionTime.append(round((random()+randint(0,19))*1000,0))
    DescriptionDwelling.append(min(round((random()+randint(0,19))*1000,0),DescriptionTime[k]))
    ReviewClicks.append(randint(0,15))
    ReviewTime.append(round((random()+randint(0,29))*1000,0))
    ReviewDwelling.append(min(round((random()+randint(0,29))*1000,0),ReviewTime[k]))
    Return.append(randint(0,1))
```

```python
Data = {'TotalTime' : TotalTime,
        'PictureClicks' : PictureClicks,
        'PictureTime' : PictureTime,
        'PictureDwelling' : PictureDwelling,
        'SummaryClicks' : SummaryClicks,
        'SummaryTime' : SummaryTime,
        'SummaryDwelling' : SummaryDwelling,
        'DescriptionClicks' : DescriptionClicks,
        'DescriptionTime' : DescriptionTime,
        'DescriptionDwelling' : DescriptionDwelling,
        'ReviewClicks' : ReviewClicks,
        'ReviewTime' : ReviewTime,
        'ReviewDwelling' : ReviewDwelling,
        'Return' : Return}

df = DataFrame(Data, columns = ['TotalTime', 'PictureClicks', 'PictureTime', 'PictureDwelling', 'SummaryClicks', 'SummaryTime', 'SummaryDwelling', 'DescriptionClicks', 'DescriptionTime', 'DescriptionDwelling', 'ReviewClicks', 'ReviewTime', 'ReviewDwelling', 'Return'])
```

```python
pd.set_option('precision', 2)
df = df.astype(int)
dfShuffle = df.sample(frac=1)
dfShuffle
```

```python
dfShuffle.to_csv("DummyData12kWithNoise.csv", index=False)
```
